```python
import numpy as np
```

```python
a=np.arange(0,5,0.2)
```

```python
plt.plot(a,a**2,'r--',label='^2')
plt.plot(a,a**2.5,'ys',label='^2.5')
plt.plot(a,a**3,'g^',label='^3')
plt.ylabel('Exponential Value')
plt.xlabel('Numbers')
plt.title('Exponential Function')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(a,a**2,'r--',label='^2',linewidth=5)
plt.plot(a,a**2.5,'ys',label='^2.5',linewidth=5)
plt.plot(a,a**3,'g^',label='^3',linewidth=5)
plt.ylabel('Exponential Value')
plt.xlabel('Numbers')
plt.title('Exponential Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def t(x):
    return np.exp(x)*np.cos(2*x*np.pi)

x1=np.arange(0,5,0.1)
x2=np.arange(0,5,0.02)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(x1,t(x1),'b')

plt.subplot(212)
plt.plot(x2,t(x2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot(x1,t(x1),'r')
plt.grid()

plt.subplot(122)
plt.plot(x2,t(x2),'y')

plt.figure(2)
plt.plot(x1,t(x1),'g')

plt.figure(3)
plt.subplot(121)
plt.plot(x1,t(x1),'r')
plt.grid()

plt.subplot(122)
plt.plot(x2,t(x2),'y')
```
