```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
#To create 1000 random observations belonging to Gaussian Distribution of mean=0 and standard deviation=1
std_normal=np.random.normal(loc=0,scale=1,size=1000)                     #normal- means gaussian distribution is being created
                                                                         #loc= mean of distribution,    scale= standard deviation of ditribution
                                                                         #size=1000 tells thaat 1000 values are being generated
```

```python
#To get all the percentiles of the distribution.
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
measurements=np.random.normal(loc=20,scale=5,size=100)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

```python
measurements=np.random.normal(loc=20,scale=5,size=1000000)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

```python
measurements=np.random.uniform(low=-1,high=1,size=1000000)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```
