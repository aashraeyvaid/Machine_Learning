```python
count,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
```

```python
count
```

```python
bin_edges
```

```python
pdf=count/(sum(count))
```

```python
pdf
```

```python
cdf=np.cumsum(pdf)
```

```python
cdf
```

```python
plt.plot(bin_edges[1:],pdf,'b-')
plt.plot(bin_edges[1:],cdf,'y-')
plt.show()
```

```python
count1,bin_edge1=np.histogram(iris_virginica['petal_length'],bins=10,density=True)
pdf1=count1/sum(count1)
cdf1=np.cumsum(pdf1)
plt.plot(bin_edge1[1:],pdf1)
plt.plot(bin_edge1[1:],cdf1)
plt.show()
```

```python
count2,bin_edge2=np.histogram(iris_versicolor['petal_length'],bins=10,density=True)
pdf2=count2/sum(count2)
cdf2=np.cumsum(pdf2)
plt.plot(bin_edge2[1:],pdf2)
plt.plot(bin_edge2[1:],cdf2)
plt.show()
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.plot(bin_edge1[1:],pdf1)
plt.plot(bin_edge1[1:],cdf1)
plt.plot(bin_edge2[1:],pdf2)
plt.plot(bin_edge2[1:],cdf2)
plt.show()
```
