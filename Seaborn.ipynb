```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=np.array([1,2,3,4,5,6])
var1=np.array([2,3,4,1,5,6])
```

```python
plt.plot(var,var1)
plt.show()
```

```python
z=pd.DataFrame({'vr':var,'vr1':var1})
```

```python
sns.lineplot(x='vr',y='vr1',data=z)
plt.show()
```

```python
data=sns.load_dataset('penguins')
```

```python
data
```

```python
sns.lineplot(x='flipper_length_mm', y='body_mass_g', data=data,hue='species')
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=data, hue='sex',size=20,style='species',palette='Accent')
```

```python
sns.distplot(data['flipper_length_mm'], kde=True)
```

```python
plt.grid()
sns.distplot(data['flipper_length_mm'],bins=np.array([170,180,190,200,210,220,230,240]),kde=True,rug=True)
```

```python
sns.barplot(x=data.island, y=data.bill_length_mm)
```

```python
order_=['Biscoe', 'Dream', 'Torgersen']
sns.barplot(x=data.island, y=data.bill_length_mm,hue=data.species, order=order_)
#The black line or dot on top of a bar shows the central value (usually the mean), and the vertical line around it shows the confidence interval — a measure of uncertainty around that estimate.
```
